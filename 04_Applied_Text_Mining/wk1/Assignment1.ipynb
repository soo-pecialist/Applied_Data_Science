# Assignment 1

In this assignment, you'll be working with messy medical data and using regex to extract relevant infromation from the data. 

Each line of the `dates.txt` file corresponds to a medical note. Each note has a date that needs to be extracted, but each date is encoded in one of many formats.

The goal of this assignment is to correctly identify all of the different date variants encoded in this dataset and to properly normalize and sort the dates. 

Here is a list of some of the variants you might encounter in this dataset:
* 04/20/2009; 04/20/09; 4/20/09; 4/3/09
* Mar-20-2009; Mar 20, 2009; March 20, 2009;  Mar. 20, 2009; Mar 20 2009;
* 20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
* Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
* Feb 2009; Sep 2009; Oct 2010
* 6/2008; 12/2009
* 2009; 2010

Once you have extracted these date patterns from the text, the next step is to sort them in ascending chronological order accoring to the following rules:
* Assume all dates in xx/xx/xx format are mm/dd/yy
* Assume all dates where year is encoded in only two digits are years from the 1900's (e.g. 1/5/89 is January 5th, 1989)
* If the day is missing (e.g. 9/2009), assume it is the first day of the month (e.g. September 1, 2009).
* If the month is missing (e.g. 2010), assume it is the first of January of that year (e.g. January 1, 2010).
* Watch out for potential typos as this is a raw, real-life derived dataset.

With these rules in mind, find the correct date in each note and return a pandas Series in chronological order of the original Series' indices.

For example if the original series was this:

    0    1999
    1    2010
    2    1978
    3    2015
    4    1985

Your function should return this:

    0    2
    1    4
    2    0
    3    1
    4    3

Your score will be calculated using [Kendall's tau](https://en.wikipedia.org/wiki/Kendall_rank_correlation_coefficient), a correlation measure for ordinal data.

*This function should return a Series of length 500 and dtype int.*

```python
import pandas as pd
import numpy as np

doc = []
with open('dates.txt') as file:
    for line in file:
        doc.append(line)

df = pd.Series(doc)
df.head(10)
```

```python
## helpers

def is_fullyear(year):
    if len(year) != 4:
        year = '19'+ year
    return year
def is_fullmonth(month):
    try:
        if len(month) == 1:
            month = '0' + month
    except:
        month = '01'
    return month
def is_fullday(day):
    try:
        if len(day) == 1:
            day = '0' + day
    except:
        day = '01'
    return day

month_map = {'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04', 'May': '05', 'Jun': '06',
             'Jul': '07', 'Aug': '08', 'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12',}

# convert to dataframe
dff = pd.DataFrame(data={'original': df.copy(), 'normalized':None})
```

```python
def type1():
    """
        04/20/2009; 04/20/09; 4/20/09; 4/3/09
        4-13-82
    """
    ## extracting date
    result1 = dff[dff['normalized'].isnull()]['original'].str.extractall(r'(?P<date>(?P<month>\d?\d)[/-](?P<day>\d?\d)[/-](?P<year>\d{2,4}))')

    ## index
    index1 = pd.Index(set(result1.index.get_level_values(0)))
    
    ## normalizing date
    result1['year'] = result1['year'].apply(is_fullyear)
    result1['month'] = result1['month'].apply(is_fullmonth)
    result1['day'] = result1['day'].apply(is_fullday)   
    result1['normalized'] = (result1['year'] + result1['month'] + result1['day'])
    
    ## extracting normalized date
    for idx in index1:
        dff.loc[idx]['normalized'] = result1.loc[idx]['normalized'].values[0]

    return
```

```python
def type2():
    """
        Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
        Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;
    """
    ## extracting date
    result2 = dff[dff['normalized'].isnull()]['original'].str.extractall(r'(?P<date>(?P<month>Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[-.\s]? ?(?P<day>\d?\d)[-.,\s] ?[a-z]*(?P<year>\d{2,4}))')

    ## index
    index2 = result2.index.get_level_values(0)    
    
    ## normalizing date
    result2['month'] = result2['month'].replace(month_map)
    result2['year'] = result2['year'].apply(is_fullyear)
    result2['month'] = result2['month'].apply(is_fullmonth)
    result2['day'] = result2['day'].apply(is_fullday)   
    result2['normalized'] = (result2['year'] + result2['month'] + result2['day'])
    
    
    ## extracting normalized date
    for idx in index2:
        dff.loc[idx]['normalized'] = result2.loc[idx]['normalized'].values[0]
        
    return
```

```python
def type3():
    """
        20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009 
        Feb 2009; Sep 2009; Oct 2010
    """
    ## extracting date
    result3 = dff[dff['normalized'].isnull()]['original'].str.extractall(r'(?P<date>(?P<day>\d?\d)? ?(?P<month>Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[,]? ?(?P<year>\d{2,4}))')

    ## index
    index3 = result3.index.get_level_values(0)
    
    ## normalizing date
    result3['month'] = result3['month'].replace(month_map)
    result3['year'] = result3['year'].apply(is_fullyear)
    result3['month'] = result3['month'].apply(is_fullmonth)
    result3['day'] = result3['day'].apply(is_fullday)   
    result3['normalized'] = (result3['year'] + result3['month'] + result3['day'])

    ## extracting normalized date
    for idx in index3:
        dff.loc[idx]['normalized'] = result3.loc[idx]['normalized'].values[0]
        
    return
```

```python
def type4():
    """
        6/2008; 12/2009 
    """
    ## extracting date
    result4 = dff[dff['normalized'].isnull()]['original'].str.extractall(r'(?P<date>[^\d+\w+/\(\)-~]?(?P<month>\d?\d)/(?P<year>\d{4}))')
    
    ## index
    index4 = result4.index.get_level_values(0)
    
    ## normalizing date
    result4['month'] = result4['month'].apply(is_fullmonth)  
    result4['normalized'] = (result4['year'] + result4['month'] + '01')
    
    ## extracting normalized date
    for idx in index4:
        dff.loc[idx]['normalized'] = result4.loc[idx]['normalized'].values[0]
    
    return
```

```python
def type5():
    """
        2009; 2010
        y1974; r1978
    """
    ## extracting date
    result5 = dff[dff['normalized'].isnull()]['original'].str.extractall(r'(?P<date>[^\d+\w+/\(\)-~]?[\w]?(?P<year>\d{4}))')
    
    ## index
    index5 = result5.index.get_level_values(0)
    
    ## normalizing date
    result5['normalized'] = (result5['year'] + '0101')
    
    ## extracting normalized date
    for idx in index5:
        dff.loc[idx]['normalized'] = result5.loc[idx]['normalized'].values[0]
        
    return
```

```python
def date_sorter():
    
    type1()
    type2()
    type3()
    type4()
    type5()
    
    dff['normalized'] = dff['normalized'].astype(str).astype(int)
    
    return dff['normalized'].argsort()
```
